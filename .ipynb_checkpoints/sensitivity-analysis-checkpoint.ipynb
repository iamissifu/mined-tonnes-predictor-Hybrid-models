```python
# 1. Import Required Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import AdaBoostRegressor
from sklearn.tree import DecisionTreeRegressor
import os
```

```python
# Load the datasets
train_df = pd.read_excel("train_data.xlsx")
test_df = pd.read_excel("test_data.xlsx")
```

```python
# 4. Define Features and Target
target_column = 'P'  # replace with actual target name if different
X_train = train_df.drop(columns=[target_column])
y_train = train_df[target_column]

X_test = test_df.drop(columns=[target_column])
y_test = test_df[target_column]
```

```python
model = AdaBoostRegressor(
    base_estimator=DecisionTreeRegressor(max_depth=4),
    n_estimators=100,
    random_state=42
)
model.fit(X_train, y_train)
```

```python
# 6. Predict on the Test Set
adaboost_predictions = model.predict(X_test)
```

```python

# 7. Compute Sensitivity Relevance Factor (SRF)
R = {}
mean_y = np.mean(adaboost_predictions)
features = X_test.columns

for feature in features:
    x = X_test[feature]
    mean_x = x.mean()
    
    numerator = np.sum((x - mean_x) * (adaboost_predictions - mean_y))
    denominator = np.sqrt(np.sum((x - mean_x)**2) * np.sum((adaboost_predictions - mean_y)**2))
    
    R[feature] = numerator / denominator if denominator != 0 else 0
```

```python
# 8. Plot the SRF Values
plt.figure(figsize=(10, 6))

# Sort the SRF values in descending order
sorted_items = sorted(R.items(), key=lambda x: abs(x[1]), reverse=True)
features_sorted = [k for k, v in sorted_items]
values_sorted = [v for k, v in sorted_items]

# Create a colormap that provides different colors for each bar
colors = plt.cm.get_cmap('tab10', len(R))  # Use 'tab10' colormap for distinct colors

# Plot horizontal bar chart with distinct colors
bars = plt.barh(range(len(R)), values_sorted, color=colors(range(len(R))))

# Set the y-ticks to be the sorted feature names
plt.yticks(range(len(R)), features_sorted)

# Label axes
plt.xlabel('Sensitivity Relevance Factor (SRF)', fontsize=12)
plt.ylabel('Features', fontsize=12)

# Set title
plt.title('Feature Sensitivity Analysis using AdaBoost', fontsize=14)

# Add gridlines
plt.grid(True, linestyle='--', alpha=0.5)

# Adjust layout for better spacing
plt.tight_layout()

# Show the plot
plt.show()
```
